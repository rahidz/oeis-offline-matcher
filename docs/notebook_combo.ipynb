# Combination search walkthrough

Demonstrates explaining a simple arithmetic progression as a linear combo of two OEIS sequences.

Requirements:
- SQLite DB built (run `oeis build-index`).
- Installed package in editable mode.

Feel free to edit the query terms / coefficients.

```python
from oeis_matcher.api import analyze_sequence

# Arithmetic progression that can be expressed as 2*A000027 (naturals) + A000012 (ones)
query = "3,5,7,9,11"
res = analyze_sequence(
    query,
    combos=10,
    combo_coeffs=(1, 2),
    combo_max_shift=1,
    combo_candidates=60,
    similarity=0,
    transform_limit=0,
)
res["combinations"][:3]
```

```python
# View expressions in both human-readable and LaTeX form
for combo in res["combinations"]:
    print(combo["expression"])
    print(combo.get("latex_expression"))
    break
```
